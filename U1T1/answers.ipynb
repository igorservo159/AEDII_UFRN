## Question 2

We need to calculate the probability of P(Not Spam|Flagged). By Bayes' Theorem:

<div style="margin-left: 40px;">
P(Not Spam|Flagged) = P(Flagged|Not Spam) * P(Not Spam) / P(Flagged)
</div>

To find P(Flagged), we use:

<div style="margin-left: 40px;">
P(Flagged) = P(Flagged|Spam) * P(Spam) + P(Flagged|Not Spam) * P(Not Spam)
</div>

Substituting the values:

<div style="margin-left: 40px;">
P(Flagged) = (0.98 * 0.6) + (0.02 * 0.4) = 0.588 + 0.008 = 0.596
</div>

Now, calculate P(Not Spam|Flagged):

<div style="margin-left: 40px;">
P(Not Spam|Flagged) = (0.02 * 0.4) / 0.596 ≈ 0.01342
</div>

Finally, the expected number of non-spam emails in the spam box is:

<div style="margin-left: 40px;">
Non-spam emails = 0.01342 * 963 ≈ 12.93
</div>

Therefore, out of 963 emails in the spam box, you should suspect that around **13 emails** are not spam.
