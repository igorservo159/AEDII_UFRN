## Question 2

We now need to find the probability \( P(\text{Not Spam} | \text{Flagged}) \). According to Bayes' Theorem:

\[
P(\text{Not Spam} | \text{Flagged}) = \frac{P(\text{Flagged} | \text{Not Spam}) \cdot P(\text{Not Spam})}{P(\text{Flagged})}
\]

\( P(\text{Flagged}) \) is calculated as:

\[
P(\text{Flagged}) = P(\text{Flagged} | \text{Spam}) \cdot P(\text{Spam}) + P(\text{Flagged} | \text{Not Spam}) \cdot P(\text{Not Spam})
\]

Substituting the values:

\[
P(\text{Flagged}) = (0.98 \times 0.6) + (0.02 \times 0.4) = 0.588 + 0.008 = 0.596
\]

Now, we can calculate \( P(\text{Not Spam} | \text{Flagged}) \):

\[
P(\text{Not Spam} | \text{Flagged}) = \frac{0.02 \times 0.4}{0.596} \approx 0.01342
\]

Finally, the expected number of non-spam emails in the spam box is:

\[
\text{Non-spam emails} = 0.01342 \times 963 \approx 12.93
\]

Therefore, out of 963 emails in the spam box, you should suspect that around **13 emails** are not spam.
