## Question 2

We need to calculate the probability of P(Not Spam|Flagged). By Bayes' Theorem:

P(Not Spam|Flagged) = P(Flagged|Not Spam) * P(Not Spam) / P(Flagged)

To find P(Flagged), we use:

P(Flagged) = P(Flagged|Spam) * P(Spam) + P(Flagged|Not Spam) * P(Not Spam)

Substituting the values:

P(Flagged) = (0.98 * 0.6) + (0.02 * 0.4) = 0.588 + 0.008 = 0.596

Now, calculate P(Not Spam|Flagged):

P(Not Spam|Flagged) = (0.02 * 0.4) / 0.596 ≈ 0.01342

Finally, the expected number of non-spam emails in the spam box is:

Non-spam emails = 0.01342 * 963 ≈ 12.93

Therefore, out of 963 emails in the spam box, you should suspect that around **13 emails** are not spam.

## Question 4

Based on the given outcome probabilities ,there is bellow the plot showing both the Probability Mass Function (PMF) and the Cumulative Distribution Function (CDF).

```python
import matplotlib.pyplot as plt
import numpy as np

# Outcomes and their corresponding probabilities
outcomes = np.array([1, 2, 3, 4, 5, 6, 7, 8])
probabilities = np.array([0.1, 0.15, 0.2, 0.21, 0.17, 0.09, 0.06, 0.02])

# Cumulative probabilities for CDF
cumulative_probabilities = np.cumsum(probabilities)

# Plotting PMF
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.bar(outcomes, probabilities, color='blue', alpha=0.6)
plt.title('Probability Mass Function (PMF)')
plt.xlabel('Outcome')
plt.ylabel('Probability')

# Plotting CDF
plt.subplot(2, 1, 2)
plt.step(outcomes, cumulative_probabilities, where='mid', color='green')
plt.title('Cumulative Distribution Function (CDF)')
plt.xlabel('Outcome')
plt.ylabel('Cumulative Probability')

plt.tight_layout()
plt.show()
```

## Question 5



```python
import pandas as pd
import numpy as np
from collections import Counter
from math import log2

data = {
    'v1': ['a', 'a', 'a', 'a', 'a', 'a', 'a', 'a', 'b', 'b', 'b', 'b', 'b', 'b', 
           'c', 'c', 'c', 'c', 'c', 'c', 'c', 'd', 'd', 'd', 'd', 'd', 'd', 'd', 'd', 
           'e', 'e', 'e', 'e', 'e', 'e', 'e', 'e', 'e', 'e', 'e', 'e', 'f', 'f', 'f', 'f', 'f'],
    'v2': ['a', 's', 'a', 'd', 'd', 'f', 'f', 'f', 'f', 'g', 'g', 'g', 'g', 'h', 
           'h', 'h', 'h', 'j', 'j', 'j', 'k', 'j', 'g', 'h', 'h', 'g', 'f', 'f', 'f', 
           'f', 'd', 'd', 's', 's', 's', 's', 's', 's', 'a', 'a', 'a', 'a', 'a', 'd', 'f', 'g']
}

df = pd.DataFrame(data)


def entropy(values):
    total = len(values)
    counts = Counter(values).values()
    return -sum((count/total) * log2(count/total) for count in counts)

def joint_entropy(v1, v2):
    total = len(v1)
    joint_counts = Counter(zip(v1, v2)).values()
    return -sum((count/total) * log2(count/total) for count in joint_counts)

H_v1 = entropy(df['v1'])
H_v2 = entropy(df['v2'])

H_v1_v2 = joint_entropy(df['v1'], df['v2'])

H_v1_given_v2 = H_v1_v2 - H_v2

savings = H_v1 - H_v1_given_v2

print(f"Entropy of v1 (H(v1)): {H_v1:.4f} bits")
print(f"Entropy of v2 (H(v2)): {H_v2:.4f} bits")
print(f"Joint entropy (H(v1, v2)): {H_v1_v2:.4f} bits")
print(f"Conditional entropy H(v1|v2): {H_v1_given_v2:.4f} bits")
print(f"Savings in encoding v1 knowing v2: {savings:.4f} bits")
```

We would need 2.53 bits to encode v1 independently and 2.86 bits to encode v2 independently.

If we knew v2, we would need 1.38 bits to encode v1 because H(v1∣v2) = H(v1,v2) − H(v2).

Therefore, we would save 1.15 bits in encoding v1 knowing v2.

## Question 2

Considering reciprocal friendship and accepted friend requests as bidirectional edges, and non-reciprocal friendships and unanswered friend requests as unidirectional edges.

```python
import matplotlib.pyplot as plt
import networkx as nx

# Criar um grafo direcionado
G = nx.DiGraph()

# Adicionar nós individualmente
G.add_node('A')
G.add_node('B')

# Adicionar múltiplos nós de uma vez
nodes_to_add = ['C', 'D', 'E', 'F', 'G', 'H']
G.add_nodes_from(nodes_to_add)

# Adicionar arestas (conexões entre nós)
# Como o grafo é direcionado, a ordem importa
G.add_edge('A', 'B')  # A considera B amigo
edges_to_add = [
    ('B', 'C'),  # B considera C amigo
    ('C', 'B'),  # C considera B amigo
    ('B', 'D'),  # B considera D amigo
    ('D', 'B'),  # D considera B amigo
    ('C', 'D'),  # C considera D amigo
    ('D', 'E'),  # D enviou solicitação para E
    ('D', 'F'),  # D enviou solicitação para F
    ('D', 'G'),  # D enviou solicitação para G
    ('G', 'D'),  # G respondeu a D
    ('D', 'H'),  # D enviou solicitação para H
    ('G', 'A'),  # G tem amizade recíproca com A
    ('A', 'G')   # A tem amizade recíproca com G
]
G.add_edges_from(edges_to_add)

# Desenhar o grafo
plt.figure(figsize=(6, 3))
pos = nx.spring_layout(G, k=0.8, seed=42)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500, 
        edge_color='gray', arrows=True, arrowstyle='->', font_size=10, font_weight='bold')

plt.title("Directed friendship graph")
plt.show()
```

## Question 3

```python
# Criar um grafo não direcionado e ponderado
G_undirected = nx.Graph()

# Adicionar os mesmos nós
G_undirected.add_nodes_from(['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'])

# Adicionar arestas com pesos
# Peso 2 se a conexão for recíproca, 1 caso contrário
edges_with_weights = [
    ('A', 'B', 1),  # A considera B amigo, não é recíproco
    ('B', 'C', 2),  # Amizade recíproca entre B e C
    ('B', 'D', 2),  # Amizade recíproca entre B e D
    ('C', 'D', 1),  # C considera D amigo, mas não é recíproco
    ('D', 'E', 1),  # D enviou solicitação para E, não respondeu
    ('D', 'F', 1),  # D enviou solicitação para F, não respondeu
    ('D', 'G', 2),  # D enviou solicitação para G, respondeu
    ('D', 'H', 1),  # D enviou solicitação para H, não respondeu
    ('G', 'A', 2)   # Amizade recíproca entre A e G
]

# Adicionar as arestas ponderadas no grafo
G_undirected.add_weighted_edges_from(edges_with_weights)

# Desenhar o grafo não direcionado e ponderado
plt.figure(figsize=(6, 4))
pos = nx.spring_layout(G_undirected, k=0.8, seed=42)
nx.draw(G_undirected, pos, with_labels=True, node_color='lightblue', node_size=500, 
        edge_color='gray', font_size=10, font_weight='bold')

# Adicionar os pesos nas arestas
edge_labels = nx.get_edge_attributes(G_undirected, 'weight')
nx.draw_networkx_edge_labels(G_undirected, pos, edge_labels=edge_labels)

plt.title("Undirected and Weighted Friendship Graph")
plt.show()
```

## Question 4

Correlation network: computed using df.corr(), where each entry represents the Pearson correlation between two vectors.

```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Dados fornecidos (você pode substituir com os dados reais do arquivo se necessário)
data = np.array([
    [1, 6, 1, 1, 1, 3, 1, 1],
    [1, 4, 4, 6, 1, 5, 3, 7],
    [1, 4, 4, 6, 1, 5, 7, 8],
    [1, 4, 6, 8, 1, 7, 8, 4],
    [1, 5, 7, 7, 2, 0, 0, 5],
    [2, 2, 5, 7, 2, 9, 9, 5],
    [3, 6, 7, 0, 3, 7, 6, 4],
    [4, 6, 9, 5, 3, 3, 3, 7],
    [4, 4, 8, 4, 4, 2, 3, 5],
    [3, 6, 7, 3, 5, 0, 3, 5],
    [5, 4, 7, 2, 5, 8, 7, 5],
    [6, 6, 9, 4, 6, 8, 7, 4],
    [7, 4, 7, 5, 7, 4, 5, 7],
    [8, 6, 0, 2, 7, 2, 4, 3],
    [9, 7, 3, 9, 8, 3, 5, 6]
])

# Nome das colunas
columns = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']

# Passo 1: Converter dados em um DataFrame
df = pd.DataFrame(data, columns=columns)

# Passo 2: Calcular a matriz de correlação
correlation_matrix = df.corr()

# Passo 3: Criar o grafo usando apenas correlações positivas
G = nx.Graph()

# Adicionar nós (as colunas do DataFrame)
G.add_nodes_from(columns)

# Adicionar arestas com correlação positiva e ignorar os loops
for i in range(len(columns)):
    for j in range(i + 1, len(columns)):  # Para evitar os loops
        corr_value = correlation_matrix.iloc[i, j]
        if corr_value > 0:  # Desenhar apenas correlações positivas
            G.add_edge(columns[i], columns[j], weight=corr_value)

# Passo 4: Desenhar a rede de correlação
pos = nx.spring_layout(G)  # Layout para a visualização da rede
plt.figure(figsize=(8, 6))

# Desenhar os nós
nx.draw_networkx_nodes(G, pos, node_size=700, node_color='lightblue')

# Desenhar as arestas com espessura proporcional ao peso
edges = G.edges(data=True)
weights = [d['weight'] * 5 for (u, v, d) in edges]  # Multiplicado para ajustar a espessura
nx.draw_networkx_edges(G, pos, edgelist=edges, width=weights)

# Desenhar os rótulos dos nós
nx.draw_networkx_labels(G, pos, font_size=12, font_color='black')

# Desenhar os rótulos das arestas (correlações)
edge_labels = {(u, v): f"{d['weight']:.2f}" for u, v, d in edges}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

# Exibir a rede de correlação
plt.title("Correlation Network (Positive Weights)")
plt.show()
```
